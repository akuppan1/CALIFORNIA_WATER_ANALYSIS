```python
# another cell
import matplotlib.pyplot as plt
%matplotlib inline
import numpy
from pylab import *
```

```python
def water_plot (water_code_wanted):
    df_cali_water_1 = df_cali_water[df_cali_water['site_code'] == water_code_wanted].copy()
    df_cali_water_2 = df_cali_water_1[['msmt_date','gwe']].copy()
    df_cali_water_2 = df_cali_water_2.sort_values(by='msmt_date')
    df_cali_water_2.plot(x='msmt_date', y='gwe', title = f'Well ID: {water_code_wanted}', legend=False, ax=ax)
    
    ax.set_xlabel('Date Range', fontweight="bold")
    ax.set_ylabel('Groundwater Elevation (inches)', fontweight="bold")
    ax.title.set_size(15)
    plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=10, ncols=1, figsize=(15,30), sharex=False, sharey=False)
for water_code, ax in zip(df_cali_water_list, axes.ravel()):
    water_plot(water_code)
```
